Next is an actual example of adding a Pokemon to your binder. I'll choose `Lucario` as the example. We use the `card_recording` function to do so. 

Once we run the line (making sure it reads in the right csv file), it will prompt us for name of a Pokemon. We type `Lucario`. For this example, we choose the following card: 

https://www.pricecharting.com/game/pokemon-mega-evolution/mega-lucario-ex-179

It is a card from the Mega Evolution set. It will then prompt us for the set name. You input your set name by picking the number associated with it on the list. If it's not listed, pick `17`, which is the Other option, and manually type in the set name. Them, it will ask if its a foil of any kind. If yes, a follow-up question asking if the card is full art is asked. This is meant for record keeping purposes if you intend to make a full art National Dex binder. Finally, it asks to provide a link to an image of the card. I used the link above (which I normally use anyways for the other set of functions since it is convenient).
